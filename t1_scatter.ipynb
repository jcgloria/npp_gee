```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def t1(t_opt):
    return 0.8 + 0.02*t_opt - 0.0005*t_opt**2

t_opt = pd.Series(range(-10, 41))
t1 = t1(t_opt)
df = pd.DataFrame({'t_opt': t_opt, 't_1': t1})
```

```python
# Setting up the plot
plt.figure(figsize=(12, 8))

# Plotting t_opt on x-axis and t_1 on y-axis as a line
plt.plot(df['t_opt'], df['t_1'], linestyle='-', marker='o')

# Setting labels, title, and legend
plt.xlabel('$t_{opt}$ (C)')
plt.ylabel('$t_1$ (C)')
plt.title('$t_{opt}$ Function for different $t_1$ values')

# Display the plot
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

```
